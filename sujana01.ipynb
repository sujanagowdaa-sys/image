```python
conversation_data = {
    "hello":"hi there! how can i help you today?",
    "hi":"hello! how are you?",
    "how are you":"I'm just a bot,but I'm doing great! how about you?",
    "what is computer vision":"Computer Vision is a branch of Artificial Intelligence (AI) that enables computers to see, understand, and interpret images and videos, similar to how humans use their eyes and brain.",
    "what is ai":"ai stands for Artificial Intellengence,AI is a field of computer science focused on creating machines or software that can perform tasks that normally require human intelligence ",
    "what is deep learning":"Deep Learning is a type of Artificial Intelligence (AI) and a subset of Machine Learning that uses artificial neural networks with many layers to learn patterns from large amounts of data.",
    "bye":"Goodbyee! have a nice day",
    "thank you":"you are welcome ! happy to help"
}
```

```python
def  sample_chat(user_input,chat_history=[]):
     user_input_lower=user_input.lower()
     response= None

     for key in conversation_data:
         if key in user_input_lower:
            response=conversation_data[key]
            break

     if not response:
      response= "sorry,i don't understand that,can you ask something else?"

     chat_history.append(("You",user_input))
     chat_history.append(("Bot",response))

     return response,chat_history


```

```python
print("Welcome to Mini conversation AI")
print("Type 'exit' or 'quit' to end the chat")

chat_history = []

while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chat ended. Goodbye!")
        break

    bot_response, chat_history = sample_chat(user_input, chat_history)
    print("Bot:", bot_response)
```
